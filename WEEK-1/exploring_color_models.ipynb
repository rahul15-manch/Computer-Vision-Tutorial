```python
import cv2
```

```python
img=cv2.imread('/Users/rahulmanchanda/Desktop/Computer Vision /images/giraffe-Kenya.png')
```

```python
# convert to grayscale
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
cv2.imwrite("grayscale_giraffe.png",gray)
```

```python
rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.imwrite("rgb_giraffe.png",rgb)
```

```python
hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
cv2.imwrite("hsv_giraffe.png",hsv)
```

```python
lab=cv2.cvtColor(img,cv2.COLOR_BGR2Lab)
cv2.imwrite("lab_giraffe.png",lab)
```
